# Linear regression using the normal equations



```
import numpy as np
np.random.seed(1)

## dataset
X1 = 500*np.random.random((250,1)) - 250

# true equation
y1 = 200 + 3.4*X1[:,0]

# noise
y1 = y1.reshape(-1,1) + np.random.normal(loc=0.0, scale=100.0, size=(250,1))
```

```
def normal_equations(X, y):
 theta = np.linalg.inv(X.T @ X) @ X.T @ y
 return theta

def predict(X_w_ones, theta):
  return X_w_ones @ theta

X_ones = np.append(np.ones((X1.shape[0],1)),X1, axis = 1)
theta = normal_equations(X_ones, y1)
y_fit = predict(X_ones, theta)
```

```
import matplotlib.pyplot as plt

y1_true = 200 + 3.4*X1[:,0]
plt.figure(figsize = (6,6))
plt.scatter(X1, y1)
plt.plot(X1, y_fit, color = "black", label= "Predicted line")
plt.plot(X1, y1_true, color = "green", label = "Actual line")
plt.legend()
plt.grid()
```
